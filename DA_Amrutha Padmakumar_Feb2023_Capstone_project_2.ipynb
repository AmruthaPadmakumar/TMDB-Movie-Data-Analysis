```python
def null_values(value):
    for i in value.columns:                             ## Performing null value treatment replacing null values with mode as the 
        mode=value[i].mode()[0]             ## column have dtype as object
        value[i].fillna(mode,inplace=True)
null_values(movie_data)      
```

## TASK.10

```python
sort=movie_data.sort_values("revenue",ascending=False).head(500)       ## top 500 movies based on revenue
```

```python
sort
```

```python
mean_budget=sort["budget"].mean()
print("mean of budget ",mean_budget)
median_budget=sort["budget"].median()
print("median of budget ",median_budget)
mode_budget=sort["budget"].mode()[0]
print("mode of budget ",mode_budget)
print("\n")
mean_revenue=sort["revenue"].mean()
print("mean of revenue ",mean_revenue)
median_revenue=sort["revenue"].mean()
print("median of revenue ",median_revenue)
mode_revenue=sort["revenue"].mode()[0]
print("mode of revenue" ,mode_revenue)
print("\n")
mean_runtime=sort["runtime"].mean()
print("mean of runtime ",mean_runtime)
median_runtime=sort["runtime"].mean()
print("median of runtime ",median_runtime)
mode_runtime=sort["runtime"].mode()[0]
print("mode of runtime",mode_runtime)
```

```python
## Perfrom outlier analysis for above three columns
```

```python
l1=["revenue","budget","runtime"]
for col in l1:
    plt.figure() 
    plt.boxplot(sort[col])
    plt.title(col)
    plt.show()
```

```python
outliers=["revenue","runtime","budget"]
for i in outliers:
    q1=sort[i].quantile(0.25)
    q3=sort[i].quantile(0.75)
    IQR=q3-q1
    uf=q3+1.5*IQR
    lf=q1-1.5*IQR
    sort=sort[(sort[i]>lf)&(sort[i]<uf)]
```

```python
## After outlier analysis
```

```python
l1=["revenue","budget","runtime"]
for col in l1:
    plt.figure() 
    plt.boxplot(sort[col])
    plt.title(col)
    plt.show()
```
